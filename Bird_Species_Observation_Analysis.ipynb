## ***1. Know Your Data***

### Import Libraries

````python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

### Dataset Loading

````python
# Load Dataset
forest='bird/Bird_monitoring_Data_FOREST.xlsx'
forest_sheets=pd.ExcelFile(forest)
forest_list=[]
for sheet in forest_sheets.sheet_names:
    df1=pd.read_excel(forest,sheet_name=sheet)
    forest_list.append(df1)
forest_df=pd.concat(forest_list,ignore_index=True)
forest_df.head()
````

````python
grass='bird/Bird_monitoring_Data_GRASSLAND.xlsx'
grass_sheets=pd.ExcelFile(grass)
grass_list=[]
for sheet in grass_sheets.sheet_names:
    df2=pd.read_excel(grass,sheet_name=sheet)
    grass_list.append(df2)
grass_df=pd.concat(grass_list,ignore_index=True)
grass_df.head()
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Merging the dataset(forest_df+grass_df)
bird_df = pd.concat([forest_df, grass_df], ignore_index=True)
````

````python
bird_df.duplicated().sum()
````

````python
#Added month column to find out seasonal analysis
bird_df['Month'] = pd.to_datetime(bird_df['Date']).dt.month
````

````python
bird_df.isnull().sum()
````

````python
#Dropping columns which are mostly empty
bird_df.drop(columns=["NPSTaxonCode","TaxonCode","Previously_Obs"], inplace=True)
````

````python
#Filling Id_method with Unknown
bird_df["ID_Method"].fillna("Unknown", inplace=True)
````
